# Strings

```python
tf_str = tf.constant("Iqman")
tf_str
```

```python
tf.strings.length(tf_str)
```

```python
tf.strings.length(tf_str).numpy()
```

```python
tf.strings.unicode_decode(tf_str,"UTF8")
```

# Array of String

```python
tf_arr = tf.constant(["Iqman","Singh","Bhatia"])
```

```python
for i in tf_arr:
    print(i.numpy(),"; Length is:",tf.strings.length(i).numpy())
```
