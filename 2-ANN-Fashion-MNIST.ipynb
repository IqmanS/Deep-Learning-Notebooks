# **Fashion MNIST Artificial Neural Network**

# Importing Libraries

```python
import numpy as np
import pandas as pd 
import tensorflow as tf
import datetime
```

# Importing Datasets

```python
train_data = pd.read_csv("/kaggle/input/fashionmnist/fashion-mnist_train.csv")
test_data = pd.read_csv("/kaggle/input/fashionmnist/fashion-mnist_test.csv")
```

```python
train_data.head()
```

```python
X_train = train_data.drop(["label"],axis=1)
y_train = train_data.label
X_test = test_data.drop(["label"],axis=1)
y_test = test_data.label
```

```python
print("No. of classes in dataset =",len(y_test.unique()))
```

# Normalizing the Images 
Changing its values from [0,255] to [0,1]

```python
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
# X_train = X_train.reshape(-1,28*28)
# X_test = X_test.reshape(-1,28*28)
```

```python
X_train.shape
```

# Building the ANN

```python
model = tf.keras.models.Sequential()
```

## Layer 1
  - Fully Connected (Dense) Hidden Layer
  - Units = 128
  - Activation = RelU
  - Input Shape = (784,) - means 784 rows would be input

```python
model.add(tf.keras.layers.Dense(units=128,activation="relu",input_shape=(784,)))
```

## Layer 2
 - Dropout Layer (It is a regularization technique where we randomly set neurons in a layer to zero/deactivated so their wts wont change. It is used to reduce chances of overfitting)
 - Fraction of the input units to drop = 0.2 

```python
model.add(tf.keras.layers.Dropout(0.2))
```

## Layer 3
 -  Output Layer
 - Units = no of classes in dataset
 - activation = softmax

```python
model.add(tf.keras.layers.Dense(units=10,activation="softmax"))
```

## Compiling Model
 - Optimizer = Adam
 - Loss = Sparse softmax (categotical) crossentropy <br><br>
Use *categorical_crossentropy* loss function when there are two or more label classes, expect labels to be provided in a one_hot representation. <br>
If you want to provide labels as integers, please use *sparse_categorical_crossentropy* loss.

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['sparse_categorical_crossentropy','accuracy'])
```

```python
model.summary()
```

# Training the ANN

```python
model.fit(X_train,y_train,epochs=20)
```

# Evaluating the ANN

```python
evals = model.evaluate(X_test,y_test)
```

```python
loss = evals[0]
sparse_categorical_crossentropy = evals[1]
accuracy = evals[2]
```

```python
print("Loss of ANN is ",loss)
print("Accuracy of ANN is ",accuracy)
```
